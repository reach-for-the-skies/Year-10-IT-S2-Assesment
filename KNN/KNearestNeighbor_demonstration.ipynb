## K-Nearest Neighbor

KNN is a type of supervised learning that can be used for classification and regretion problems.

```python
#import all required libraries

#Sklearn provides many useful tools for machine learning
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model, preprocessing

#pandas can analyse, clean, explor and manipulate data which
#is important for getting data
import pandas as pd

#numpy can perform complex mathemtaical operations on arrays
import numpy as np
```

```python
#read in the data given that it is in the same directory
data = pd.read_csv("car.data")

#convert non-numerical to numerical data
le = preprocessing.LabelEncoder()

#sets the different attributes
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))
```

```python
predict = "class"

#set attributes, zip() combines it into a tuple list
X = list(zip(buying, maint, door, persons, lug_boot, safety))
y = list(cls)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)


#choose hyperparameter
model = KNeighborsClassifier(n_neighbors=9)


#fit the model
model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print(acc)

predicted = model.predict(x_test)
names = ["unacc", "acc", "good", "vgood"]
```

```python
for x in range(len(predicted)):
    print("Predicted: ", names[predicted[x]], "Data: ", x_test[x], "Actual: ", names[y_test[x]])
    n = model.kneighbors([x_test[x]], 9, True)
    print("N: ", n)
```
